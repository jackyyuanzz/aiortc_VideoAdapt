```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rate_hist = np.loadtxt('bitrate_record.txt',delimiter=',')
```

```python
rate_hist = np.loadtxt('bitrate_record0.txt',delimiter=',')
time = rate_hist[:,0] - rate_hist[0,0]
plt.figure(figsize=(15,10))
plt.plot(time, rate_hist[:,1]/1000, label='target_bitrate (kbps)')
plt.plot(time, rate_hist[:,2]/1000, label='estimated throughput (kbps)')
plt.grid()
plt.legend(fontsize=12)
plt.xlabel('Time (s)', fontsize=16); plt.ylabel('Rate (Kbps)', fontsize=16)
```
